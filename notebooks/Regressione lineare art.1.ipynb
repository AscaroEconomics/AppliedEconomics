Regressione Lineare e OLS: un'applicazione pratica

```python
from matplotlib.pyplot import subplots
import matplotlib.pyplot as plt
import wooldridge
import statsmodels.api as sm
import seaborn as sns
import numpy as np
```

Vediamo come si presenta il dataset visualizzando le prime 5 righe:

```python
df = wooldridge.data("wage1")

df.head()
```

Visualizziamo alcune informazioni sulle variabili (media, deviazione standard, minimo, massimo ecc.)

```python
df.describe()
```

Visualizziamo tramite uno scatterplot la correlazione tra Istruzione e Salario

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='educ', y='wage')
plt.title('Salario Vs Istruzione')
plt.xlabel('Anni di istruzione')
plt.ylabel('Salario')
plt.show()
```

Stimiamo i parametri della regressione lineare semplice utilizzando il modello OLS

```python
x = df['educ']
X = sm.add_constant(x)
y = df['wage']
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
beta_0, beta_1 = model.params

print(beta_0, beta_1)
```

```python
y_hat = model.fittedvalues
```

```python
y_hat
```

```python
e = y - y_hat

e
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='educ', y='wage')
sns.lineplot(data = df, x = x, y = beta_0 + beta_1*x, palette = "Set1")
sns.scatterplot(data = df, x = x, y = beta_0 + beta_1*x, marker = "s")
plt.vlines(x, y, y_hat)
plt.title('Salario Vs Istruzione')
plt.xlabel('Anni di istruzione')
plt.ylabel('Salario')
plt.show()
```

```python
# 5️⃣ OLS con controlli aggiuntivi
# ----------------------------
# Aggiungiamo controlli: esperienza lavorativa (exper), anni con l'attuale datore di lavoro (tenure), sposato/a
X_ctrl = df[['educ', 'exper', 'tenure', 'married']]
X_ctrl = sm.add_constant(X_ctrl)
```

```python
model_ctrl = sm.OLS(y, X_ctrl).fit()
print(model_ctrl.summary())
```
